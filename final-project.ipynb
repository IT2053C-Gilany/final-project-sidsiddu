## üìä STEP 3 - File Loading with Exception Handling

### Importing Libraries & File Loading with Exception Handling


```python
# üöÄ Importing Libraries
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from typing import List, Dict, Tuple
from scipy.stats import ttest_ind, pearsonr, chi2_contingency

# üé® Visual and display settings
sns.set_theme(style="whitegrid", palette="Set2")
pd.set_option("display.max_columns", None)

print("‚úÖ Libraries loaded successfully!")

# üìÅ Function: Load CSV with custom error handling


def load_csv(path: Path) -> pd.DataFrame:
  """
  Load a CSV file with robust exception handling.

  Args:
      path (Path): Path to the dataset file.

  Returns:
      pd.DataFrame: Loaded dataframe.

  Raises:
      FileNotFoundError: If the file does not exist.
      ValueError: If the file cannot be parsed due to format issues.
  """
  try:
    df = pd.read_csv(path, na_values="?")
  except FileNotFoundError as e:
    raise FileNotFoundError(
        f"‚ùå ERROR: File '{path}' was not found. Please check the data/ folder."
    ) from e
  except pd.errors.ParserError as e:
    raise ValueError(
        f"‚ùå ERROR: Parsing failed for '{path.name}'. Check delimiter or malformed rows."
    ) from e
  else:
    print(f"üìÅ Successfully loaded dataset: {path.name}")
    return df


# üìä Loading the Data
data_path = Path("data/imports-85.data")
df_raw = load_csv(data_path)

# üëÄ Quick Look at the Dataset
df_raw.head()
```

## üìä STEP 4 - DATA CLEANING & PYTHON STRUCTURES

```python
# ===============================================================
# üßπ STEP 4 ‚Äî Data Cleaning & Python Structures (Safe & Independent)
# ===============================================================

print("üîß Starting data cleaning and preparation...")

df = df_raw.copy()

# ---------------------------------------------
# üõë SAFETY CHECK ‚Äî Ensure column names exist
# ---------------------------------------------

expected_columns = [
    "symboling", "normalized_losses", "make", "fuel_type", "aspiration",
    "num_doors", "body_style", "drive_wheels", "engine_location",
    "wheelbase", "length", "width", "height", "curb_weight",
    "engine_type", "num_cylinders", "engine_size", "fuel_system",
    "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
    "city_mpg", "highway_mpg", "price"
]

if list(df.columns) != expected_columns:
  print("‚ö†Ô∏è Column names missing or incorrect ‚Äî applying correct column names now.")
  df.columns = expected_columns
else:
  print("‚úÖ Column names already correctly assigned!")

# ---------------------------------------------
# üî† Python Structures (Required by assignment)
# ---------------------------------------------

# üü¶ SET: unique body styles
unique_body_styles = set(df["body_style"])
print(f"üü¶ Unique body styles found: {unique_body_styles}")

# üü• DICT: mapping fuel types
fuel_map: Dict[str, str] = {
    "gas": "Gasoline",
    "diesel": "Diesel"
}
df["fuel_type_label"] = df["fuel_type"].map(fuel_map)

# üü© TUPLE: grouped dimension columns
dimension_cols: Tuple[str, str, str] = ("length", "width", "height")

# üü® LIST: engine-related numeric features
engine_features: List[str] = ["engine_size", "horsepower", "peak_rpm"]

print("üîß Applied list, tuple, set, and dict successfully!")

# ---------------------------------------------
# üî¢ Clean Numeric Columns
# ---------------------------------------------

numeric_cols = [
    "normalized_losses", "wheelbase", "length", "width", "height",
    "curb_weight", "engine_size", "bore", "stroke",
    "compression_ratio", "horsepower", "peak_rpm",
    "city_mpg", "highway_mpg", "price"
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

print("üî¢ Numeric cleaning completed!")


# ---------------------------------------------
# üóëÔ∏è Remove rows without price
# ---------------------------------------------

rows_before = df.shape[0]
df = df.dropna(subset=["price"])
rows_after = df.shape[0]

print(f"üóëÔ∏è Removed {rows_before - rows_after} rows with missing price values.")

# üëÄ Final preview
df.head()
```

### ü™ú STEP 6.4 ‚Äî T-Test (Gas vs Diesel Price)

```python
# ===============================================================
# üß™ T-Test: Gas vs Diesel Vehicle Prices
# ===============================================================

gas_prices = df[df["fuel_type"] == "gas"]["price"]
diesel_prices = df[df["fuel_type"] == "diesel"]["price"]

t_stat, p_val = ttest_ind(gas_prices, diesel_prices, equal_var=False)

print("üß™ Welch T-Test: Gas vs Diesel Prices")
print("-------------------------------------")
print(f"üìå t-statistic: {t_stat:.3f}")
print(f"üìå p-value: {p_val:.5f}")

# Effect size (Cohen's d)
mean_diff = gas_prices.mean() - diesel_prices.mean()
pooled_sd = np.sqrt((gas_prices.var() + diesel_prices.var()) / 2)
effect_size = mean_diff / pooled_sd

print(f"üìè Effect Size (Cohen's d): {effect_size:.3f}")
```

### ü™ú STEP 6.5 ‚Äî Interpretation

### üîç Interpretation: Gas vs Diesel Pricing

>- The p-value indicates whether gas and diesel vehicle prices differ significantly.
>- If p < 0.05 ‚Üí reject the null hypothesis ‚Üí prices differ meaningfully.
>- Cohen‚Äôs d shows effect size:
>>- 0.2 = small
>>- 0.5 = medium
>>- 0.8+ = large

Diesel vehicles often cost more due to efficiency and turbocharging technology.
This test formally quantifies that difference.

### ü™ú STEP 6.6 ‚Äî Chi-Square Test (Fuel Type √ó Aspiration)

```python
# ===============================================================
# üî¢ Chi-Square Test: Fuel Type √ó Aspiration
# ===============================================================

contingency = pd.crosstab(df["fuel_type"], df["aspiration"])
chi2, p, dof, expected = chi2_contingency(contingency)

print("üî¢ Chi-Square Test for Categorical Association")
print("---------------------------------------------")
print("üìä Contingency Table:")
display(contingency)

print(f"\nüìå Chi-square statistic: {chi2:.3f}")
print(f"üìå p-value: {p:.5f}")
print(f"üìå Degrees of freedom: {dof}")
```

### ü™ú STEP 6.7 ‚Äî Interpretation

### üîç Interpretation: Categorical Association

The chi-square test checks whether fuel type (gas/diesel) is related to aspiration (std/turbo).

>- If p < 0.05 ‚Üí significant association
>- If p ‚â• 0.05 ‚Üí no meaningful relationship

Aspiration often influences performance, and diesel cars commonly use turbochargers.
The chi-square test quantifies whether this trend holds statistically.

### ü™ú STEP 6.8 ‚Äî Correlation + 95% CI (Horsepower vs Price)

```python
# ===============================================================
# üìà Correlation Between Horsepower and Price (with 95% CI)
# ===============================================================

corr, p_val = pearsonr(df["horsepower"], df["price"])

# Fisher Transformation for CI
z = np.arctanh(corr)
se = 1 / np.sqrt(df.shape[0] - 3)
ci_low = np.tanh(z - 1.96 * se)
ci_high = np.tanh(z + 1.96 * se)

print("üìà Pearson Correlation: Horsepower vs Price")
print("-------------------------------------------")
print(f"üîó Correlation: {corr:.3f}")
print(f"üìå p-value: {p_val:.5f}")
print(f"üìâ 95% CI: ({ci_low:.3f}, {ci_high:.3f})")
```

### ü™ú STEP 6.9 ‚Äî Interpretation

### üîç Interpretation: Horsepower ‚Üí Price Relationship

>- Strong positive correlation suggests powerful engines tend to increase price.
>- Confidence intervals help confirm reliability:
>>- Narrow CI ‚Üí stable estimate
>>- Wide CI ‚Üí uncertain relationship
>- The p-value shows whether the correlation is statistically significant.

This validates earlier visual trends from the scatter plot + regression.
